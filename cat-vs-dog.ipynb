```python
from keras import Sequential
from keras import layers
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import os
import numpy as np
from glob import glob
from PIL import Image
```

```python
# import do dataset
dataset = 'PetImages'
```

```python
# ajustando as imagens
img_size = (64,64)
batch_size = 32
```

```python
# listas de classes
class_name = sorted(os.listdir(dataset))
print(f'classes detectadas {class_name}')
```

```python
# carregar imagens
imagens = []
labels = []
```

```python
for label, class_name in enumerate(class_name):
    class_dir = os.path.join(dataset, class_name)
    images_path = glob(os.path.join(class_dir,'*.jpg'))

    for img_path in images_path:
        img = Image.open(img_path).convert('RGB')
        img = img.resize(img_size)
        imagens.append(np.array(img))
        labels.append(label)
```

```python
imagens = np.array(imagens) / 255.0
labels = np.array(labels)
```

```python
X_train,X_test, y_train, y_test = train_test_split(imagens, labels, test_size = 0.2,
                                                   random_state = 0, stratify = labels)
X_test
```

```python
modelo = Sequential()
```

```python
modelo.add(layers.InputLayer(input_shape = (64,64,3)))
```

```python
modelo.add(layers.Conv2D(32,(3,3), activation = 'relu', padding = 'same'))
modelo.add(layers.MaxPooling2D((2,2)))
```

```python
modelo.add(layers.Conv2D(64,(3,3), activation = 'relu', padding = 'same'))
modelo.add(layers.MaxPooling2D((2,2)))
```

```python
modelo.add(layers.Conv2D(128,(3,3), activation = 'relu', padding = 'same'))
modelo.add(layers.MaxPooling2D((2,2)))
```

```python
# transformar a saida em um vetor
modelo.add(layers.Flatten())
```

```python
# camada densa
modelo.add(layers.Dense(128, activation = 'relu'))
modelo.add(layers.Dropout(0.5))
```

```python
modelo.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
modelo.compile(optimizer = "adam",
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

```python
modelo.fit(X_train, y_train, epochs = 10, batch_size = 32, validation_data=(X_test, y_test))
```
